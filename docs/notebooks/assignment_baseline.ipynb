# Assignment Baseline

This notebook solves a one-to-one robot–task assignment problem using
`scipy.optimize.linear_sum_assignment`.

The experiment verifies the mathematical formulation in
`docs/assignment_problem.md`.

```python
import numpy as np
from scipy.optimize import linear_sum_assignment

cost_matrix = np.array([
    [4, 7, 6],
    [5, 3, 8],
    [7, 6, 4],
])

robot_indices, task_indices = linear_sum_assignment(cost_matrix)

total_cost = cost_matrix[robot_indices, task_indices].sum()

for robot, task in zip(robot_indices, task_indices):
    print(
        f"Robot R{robot + 1} -> Task T{task + 1}, "
        f"cost = {cost_matrix[robot, task]}"
    )

print(f"Total cost: {total_cost}")
```
